```python
# imports

import matplotlib.pyplot as plt
import umap
import matplotlib.colors as mcolors
import random
import numpy
```

```python
# reads the embedded units of text from file
def read_text_units_embedded(text_unit):
    try:
        with open(f"../data/{text_unit}/{text_unit}_embedded.txt", "r") as file:
            lines = file.readlines()
            return lines
    except Exception as ex:
        raise
```

```python
chunk_vectors_str = read_text_units_embedded("chunks")

chunk_vectors_768 =[]
for chunk_str in chunk_vectors_str:
    chunk_vectors_768.append(numpy.fromstring(chunk_str, sep=' '))


sen_vectors_str = read_text_units_embedded("sens")
sen_vectors_768 =[]
for sen_str in sen_vectors_str:
    sen_vectors_768.append(numpy.fromstring(sen_str, sep=' '))
```

```python
print(len(chunk_vectors_768))
print(len(sen_vectors_768))
```

```python
# transforming the verctor dimension 768 -> 2

# initialization of the reducer
reducer = umap.UMAP(n_components=2) # reducer to transform  to 2 dimension

chunk_vectors_2 = reducer.fit_transform(chunk_vectors_768)
sen_vectors_2 = reducer.fit_transform(sen_vectors_768)
```

```python
plt.scatter(chunk_vectors_2[:, 0], chunk_vectors_2[:, 1], s=5) # [:,0] kijeloli az osszes sort, es veszi a 0. indexu oszlopot
plt.title('projection of chunk vectors')
plt.show()
```

```python
plt.scatter(sen_vectors_2[:, 0], sen_vectors_2[:, 1], s=5) # [:,0] kijeloli az osszes sort, es veszi a 0. indexu oszlopot
plt.title('projection of sentence vectors')
plt.show()
```
